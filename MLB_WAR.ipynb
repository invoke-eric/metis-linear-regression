# Imports and Functions

```python
import requests
from bs4 import BeautifulSoup as bs
from bs4 import Comment
import pandas as pd
import numpy as np
```

```python
def create_player_df(url):
    url = url
    page = requests.get(url).text
    soup = bs(page)
    
    players = soup.find_all(attrs={'data-stat': 'player'})
    
    name_list = []
    for i in players:
        if i.text != 'Name':
            name_list.append(i.text)
    
    name_additional_list =[]
    for i in players:
        try:
            name_additional_list.append(i['data-append-csv'])
        except:
            pass
    
    age_list = []
    age = soup.find_all(attrs={'data-stat': 'age'})
    for i in age:
        if i.text != 'Age':
            age_list.append(i.text)
            
    team_list = []
    team = soup.find_all(attrs={'data-stat': 'team_ID'})
    for i in team:
        if i.text != 'Tm':
            team_list.append(i.text)
            
    
    lg_list = []
    lg = soup.find_all(attrs={'data-stat': 'lg_ID'})
    for i in lg:
        if i.text != 'Lg':
            lg_list.append(i.text)
    
    games_list = []
    games = soup.find_all(attrs={'data-stat': 'G'})
    for i in games:
        if i.text != 'G':
            games_list.append(i.text)
    
    AB_list = []
    AB =  soup.find_all(attrs={'data-stat': 'AB'})
    for i in AB:
        if i.text != 'AB':
            AB_list.append(i.text)
            
    R_list = []
    R = soup.find_all(attrs={'data-stat': 'R'})
    for i in R:
        if i.text != 'R':
            R_list.append(i.text)
            
    H_list = []
    H = soup.find_all(attrs={'data-stat': 'H'})
    for i in H:
        if i.text != 'H':
            H_list.append(i.text)
    
    PA_list = []
    PA = soup.find_all(attrs={'data-stat': 'PA'})
    for i in PA:
        if i.text != 'PA':
            PA_list.append(i.text)
            
    doubles_list = []
    doubles = soup.find_all(attrs={'data-stat': '2B'})
    for i in doubles:
        if i.text != '2B':
            doubles_list.append(i.text)
    
    
    triples_list = []
    triples = soup.find_all(attrs={'data-stat': '3B'})
    for i in triples:
        if i.text != '3B':
            triples_list.append(i.text)
            
    HR_list = []
    HR = soup.find_all(attrs={'data-stat': 'HR'})
    for i in HR:
        if i.text != 'HR':
            HR_list.append(i.text)
    
    RBI_list = []
    RBI = soup.find_all(attrs={'data-stat': 'RBI'})
    for i in RBI:
        if i.text != 'RBI':
            RBI_list.append(i.text)
            
    SB_list = []
    SB = soup.find_all(attrs={'data-stat': 'SB'})
    for i in SB:
        if i.text != 'SB':
            SB_list.append(i.text)
            
    CS_list = []
    CS = soup.find_all(attrs={'data-stat': 'CS'})
    for i in CS:
        if i.text != 'CS':
            CS_list.append(i.text)
    
    BB_list = []
    BB = soup.find_all(attrs={'data-stat': 'BB'})
    for i in BB:
        if i.text != 'BB':
            BB_list.append(i.text)
            
    SO_list = []
    SO = soup.find_all(attrs={'data-stat': 'SO'})
    for i in SO:
        if i.text != 'SO':
            SO_list.append(i.text)
            
    BA_list = []
    BA = soup.find_all(attrs={'data-stat': 'batting_avg'})
    for i in BA:
        if i.text != 'BA':
            BA_list.append(i.text)
    
    OBP_list = []
    OBP = soup.find_all(attrs={'data-stat': 'onbase_perc'})
    for i in OBP:
        if i.text != 'OBP':
            OBP_list.append(i.text)
            
    SLG_list = []
    SLG = soup.find_all(attrs={'data-stat': 'slugging_perc'})
    for i in SLG:
        if i.text != 'SLG':
            SLG_list.append(i.text)
    
    OPS_list = []
    OPS = soup.find_all(attrs={'data-stat': 'onbase_plus_slugging'})
    for i in OPS:
        if i.text != 'OPS':
            OPS_list.append(i.text)
    
    OPSplus_list = []
    OPSplus = soup.find_all(attrs={'data-stat': 'onbase_plus_slugging_plus'})
    for i in OPSplus:
        if i.text != 'OPS+':
            OPSplus_list.append(i.text)
    
    TB_list = []
    TB = soup.find_all(attrs={'data-stat': 'TB'})
    for i in TB:
        if i.text != 'TB':
            TB_list.append(i.text)
    
    GIDP_list = []
    GIDP = soup.find_all(attrs={'data-stat': 'GIDP'})
    for i in GIDP:
        if i.text != 'GDP':
            GIDP_list.append(i.text)
    
    HBP_list = []
    HBP = soup.find_all(attrs={'data-stat': 'HBP'})
    for i in HBP:
        if i.text != 'HBP':
            HBP_list.append(i.text)
            
    SH_list = []
    SH = soup.find_all(attrs={'data-stat': 'SH'})
    for i in SH:
        if i.text != 'SH':
            SH_list.append(i.text)
    
    SF_list = []
    SF = soup.find_all(attrs={'data-stat': 'SF'})
    for i in SF:
        if i.text != 'SF':
            SF_list.append(i.text)
            
    IBB_list = []
    IBB = soup.find_all(attrs={'data-stat': 'IBB'})
    for i in IBB:
        if i.text != 'IBB':
            IBB_list.append(i.text)
    
    zipped = list(zip(name_list, name_additional_list, age_list, team_list, lg_list, PA_list, AB_list, R_list, H_list, doubles_list, triples_list,
                 HR_list, RBI_list, SB_list, CS_list, BB_list, SO_list, BA_list, OBP_list, OPS_list, OPSplus_list,
                 TB_list, GIDP_list, HBP_list, SH_list, SF_list, IBB_list))
    c = ['Name', 'Name-additional', 'Age', 'Tm', 'Lg', 'PA', 'AB', 'R', 'H', '2B', '3B', 'HR', 'RBI',
              'SB', 'CS', 'BB', 'SO', 'BA', 'OBP', 'OPS', 'OPS+', 'TB', 'GIDP', 'HBP', 'SH', 'SF', 'IBB']
    df = pd.DataFrame(zipped, columns=c)
    new_df = df.replace(r'^\s*$', np.nan, regex=True)
    new_df = new_df.dropna()
    convert_dict = {'Name': object,
                'Name-additional': object,
                'Age': int,
                'Tm': object,
                'Lg': object,
                'PA': int,
                'AB': int,
                'R': int,
                'H': int,
                '2B': int,
                '3B': int,
                'HR': int,
                'RBI': int,
                'SB': int,
                'CS': int,
                'BB': int,
                'SO': int,
                'BA': float,
                'OBP': float,
                'OPS': float,
                'OPS+': float,
                'TB': int,
                'GIDP': int,
                'HBP': int,
                'SH': int,
                'SF': int,
                'IBB': int
                }
    # thanks stackoverflow https://stackoverflow.com/questions/63728644/convert-multiple-columns-from-string-to-float-and-int
    new_df = new_df.astype(convert_dict)
    return(new_df)
    
```

```python
def create_WAR_df(url):
    url = url
    page = requests.get(url).text
    soup = bs(page)
    
    players = soup.find_all(attrs={'data-stat': 'player'})
    
    WAR_name_additional_list = []
    for i in players:
        try:
            WAR_name_additional_list.append(i['data-append-csv'])
        except:
            pass
    
    WAR_list = []
    WAR = soup.find_all(attrs={'data-stat': 'WAR'})
    for i in WAR:
        if i.text != 'WAR':
            WAR_list.append(i.text)
    
    WAR_df = pd.DataFrame(list(zip(WAR_list, WAR_name_additional_list)), columns=['WAR', 'Name-additional'])
    new_df = WAR_df.replace(r'^\s*$', np.nan, regex=True) # https://sparkbyexamples.com/pandas/pandas-replace-blank-values-with-nan/
    new_df = new_df.dropna()
    convert_dict = {'WAR': float,
                'Name-additional': object,}
    new_df = new_df.astype(convert_dict)
    return(new_df)
```

```python
def add_WAR_df(batting_df, value_df, year):
    batting_df['Year'] = year
    waa_df = value_df[['Name-additional', 'WAR']]
    
    #the Name-additional field is a UID per player to account for different players with the same name
    joined = batting_df.set_index('Name-additional').join(waa_df.set_index('Name-additional'))
    joined = joined.dropna()
    
    #reset the index because the prior index doesn't matter
    joined_new = joined.reset_index()
    joined_new['Name-year'] = joined_new['Name-additional'] + joined_new['Year'].astype(str)
    return(joined_new)
```

```python
def create_curr_WAR_df(url):
    url = url
    page = requests.get(url).text
    soup = bs(page)
    
    players = soup.find_all(attrs={'data-stat': 'player'})
    
    WAR_name_additional_list = []
    for i in players:
        try:
            WAR_name_additional_list.append(i['data-append-csv'])
        except:
            pass
    
    WAR_list = []
    WAR = soup.find_all(attrs={'data-stat': 'WAR'})
    for i in WAR:
        if i.text != 'WAR':
            WAR_list.append(i.text)
    
    WAR_df = pd.DataFrame(list(zip(WAR_list, WAR_name_additional_list)), columns=['curr_WAR', 'Name-additional'])
    new_df = WAR_df.replace(r'^\s*$', np.nan, regex=True)
    new_df = new_df.dropna()
    convert_dict = {'curr_WAR': float,
                'Name-additional': object,}
    new_df = new_df.astype(convert_dict)
    return(new_df)
```

```python
def add_curr_WAR_df(batting_df, value_df, year):
    batting_df['Year'] = year
    waa_df = value_df[['Name-additional', 'curr_WAR']]
    
    #the Name-additional field is a UID per player to account for different players with the same name
    joined = batting_df.set_index('Name-additional').join(waa_df.set_index('Name-additional'))
    joined = joined.dropna()
    
    #reset the index because the prior index doesn't matter
    joined_new = joined.reset_index()
    joined_new['Name-year'] = joined_new['Name-additional'] + joined_new['Year'].astype(str)
    return(joined_new)
```

```python
def deduplicate_dataframe(input_df):
    df = input_df[input_df['PA'] > 120]
    
    #select players who played for multiple teams across NL and AL
    duplicatePlayers = df[(df['Tm'] == 'TOT') & (df['Lg'] == 'MLB')]
    duplicatePlayers = duplicatePlayers.drop_duplicates()  

    duplicate_nl_al = df[(df['Tm'] == 'TOT') & ((df['Lg'] == 'NL') | (df['Lg'] == 'AL'))]
    
    new_df = df.copy()
    new_df['Duplicate1'] = df['Name-additional'].isin(duplicatePlayers['Name-additional'])
    new_df['Duplicate2'] = df['Name-additional'].isin(duplicate_nl_al['Name-additional'])

    
    #drop based on the masks
    dropped=new_df[new_df.Duplicate1 != True]
    dropped2=dropped[dropped.Duplicate2 != True]    
    dropped3 = dropped2.drop(columns=['Duplicate1', 'Duplicate2'])
    
    #Add the duplicate players back in, this time only adding a single row (see the team 'TOT')
    dedup_df = pd.concat([dropped3, duplicate_nl_al, duplicatePlayers])
    dedup_df = dedup_df.dropna() #drop nulls, should be minimal
    dedup_df = dedup_df.drop_duplicates()
    
    return(dedup_df)
```

```python
def create_joined_df(batting_url, WAR_url, curr_WAR_url, year):
    batting_df = create_player_df(batting_url)
    WAR_df = create_WAR_df(WAR_url)
    curr_WAR_df = create_curr_WAR_df(curr_WAR_url)
    batting_df_dedup = deduplicate_dataframe(batting_df)
    joined_df = add_WAR_df(batting_df_dedup, WAR_df, year)
    joined_df_2 = add_curr_WAR_df(joined_df, curr_WAR_df, year)
    return(joined_df_2)
```
